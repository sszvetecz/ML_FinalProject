```python
import pandas as pd
import numpy as np
import pyreadr

from data_generation_simulation import simulate_chemo_proteinlevel_nonparametric
```

```python
import pyreadr

# read bio data interaction .RDS template and convert to Python dict
bio_data_template1 = pyreadr.read_r("bio_data_template1.RDS")  
bio_data_template1 = bio_data_template1[None]  

# set data frames to python dictionary 
template = {
    "strong_interaction": bio_data_template1["strong_interaction"],
    "weak_interaction": bio_data_template1["weak_interaction"],
    "no_interaction": bio_data_template1["no_interaction"]
}
```

```python
# generate testing data 
# Dose values to match the allowed concentrations
concentrations = [0, 1, 3, 10, 30, 100, 300, 1000, 3000]

# strong interacting protein template (i.e. deccreasing relationship with drug that reaches near 0) 
strong_interaction = pd.DataFrame({
    "dose": concentrations,
    "LogIntensities": [5.4, 5.28, 4.75, 
})

# weak interacting protein template (i.e. decreasing relationship with drug that does no reach 0) 
weak_interaction = pd.DataFrame({
    "dose": concentrations,
    "LogIntensities": np.linspace(5.5, 4.5, len(concentrations)) + np.random.normal(0, 0.05, len(concentrations))
})

# no interacting protein template (i.e. flat line with variation) 
no_interaction = pd.DataFrame({
    "dose": concentrations,
    "LogIntensities": np.full(len(concentrations), 5.5) + np.random.normal(0, 0.05, len(concentrations))
})

# Combine into a template dictionary
template = {
    "strong_interaction": strong_interaction,
    "weak_interaction": weak_interaction,
    "no_interaction": no_interaction
}
```

```python
sim_data = simulate_chemo_proteinlevel_nonparametric(
    N_proteins=30,
    TP=0.2,
    TW=0.3,
    TN=0.5,
    concentrations=concentrations,
    rep=3,
    seed=42,
    var_tech=0.25,
    template=template,
    outlier_prob=0.05
)

sim_data.head()
```
